```python
from matplotlib.ticker import MultipleLocator
def make_patch_spines_invisible(ax):
    ax.set_frame_on(True)
    ax.patch.set_visible(False)
    for sp in ax.spines.values():
        sp.set_visible(False)

fig, host = plt.subplots(figsize=(5, 7))
fig.subplots_adjust(right=0.95)
#twinx() creates another axes sharing the x axis. We do this twice for 
#red and blue extra axes
par1 = host.twiny()
par2 = host.twiny()
# Offset the right spine of par2.  The ticks and label have already been
# placed on the right by twinx above.
par2.spines["top"].set_position(("axes", 1.1))
# Having been created by twinx, par2 has its frame off, so the line of its
# detached spine is invisible.  First, activate the frame but make the patch
# and spines invisible.
make_patch_spines_invisible(par2)
# Second, show the right spine.
par2.spines["top"].set_visible(True)
sali, = host.plot(data[:,4],-data[:,0],linewidth=1,marker='o',markersize=0.5, color='blue')
tempi, =par1.plot(data[:,1],-data[:,0],linewidth=1,marker='o',markersize=0.5, color='red')
oxi, =par2.plot(data[:,2],-data[:,0],linewidth=1,marker='o',markersize=0.5, color='black')

host.set_xlim(34.5,35)
host.set_xlabel('Salinidad')
host.set_ylabel('Profundidad')
par1.set_xlabel('Temperatura')
par2.set_xlabel('Oxigeno')

host.xaxis.label.set_color('blue')
par1.xaxis.label.set_color('red')
par2.xaxis.label.set_color('black')

tkw = dict(size=6, width=1.5)
minorLocatory = MultipleLocator(100)
minorLocatorx = MultipleLocator(10)
host.yaxis.set_minor_locator(minorLocatory)
host.xaxis.set_minor_locator(minorLocatorx)

minorLocatory = MultipleLocator(10)
par1.xaxis.set_minor_locator(minorLocatory)

minorLocatory = MultipleLocator(100)
par2.xaxis.set_minor_locator(minorLocatory)

host.tick_params(axis='x', colors='blue', **tkw)
par1.tick_params(axis='x', colors='red', **tkw)
par2.tick_params(axis='x', colors='black', **tkw)
host.tick_params(direction='in', **tkw)
plt.text(0.1, 0.5, 'Linea de Máncora 211105-04')
#plt.text(0.5, 0.5, '5ºS - 81º25.02W')

lines = [sali, tempi, oxi]

host.legend(lines, [l.get_label() for l in lines],fontsize=12, frameon=False,loc=1, bbox_to_anchor=(1.02,0.9))

# plt.savefig('/content/drive/MyDrive/Santa_Monica/'+' -linea_mancora_22'+'.png',
#     format='png', dpi=600, transparent=False, bbox_inches='tight')
```

```python
def read_xyz(inputfile):
  """ Reads xyz files """
  a2s  = np.loadtxt(inputfile, skiprows=2, usecols=[0], dtype=str)
  a2xyz = np.loadtxt(inputfile, skiprows=2, usecols=[1,2,3])
  assert len(a2s)==len(a2xyz)
  return a2s,a2xyz 
```
